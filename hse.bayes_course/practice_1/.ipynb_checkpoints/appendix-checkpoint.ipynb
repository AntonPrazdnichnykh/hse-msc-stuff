# Матожидания и дисперсии

```python
def mean_var(p_distr, params, model):
    probs, vals = p_distr(params, model)
    m = (vals * probs).sum()
    d = (vals**2 * probs).sum() - m**2
    return m, d
```

```python
#a
mean_var(pa, params, 3)
```

```python
#b
mean_var(pb, params, 3)
```

```python
#c3
mean_var(pc, params, model=3)
```

```python
#c4
mean_var(pc, params, model=4)
```

```python
#d3
mean_var(pd, params, model=3)
```

```python
#d4
mean_var(pd, params, model=4)
```
